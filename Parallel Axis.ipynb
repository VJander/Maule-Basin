```python
import plotly.graph_objects as go

import pandas as pd

df = pd.read_excel('Average_Results.xls',sheet_name='Average_Results', usecols='B:J')

fig = go.Figure(data=
    go.Parcoords(
        labelfont=dict(size=9.5),
        tickfont=dict(size=12),
        rangefont=dict(size=12),
        labelside='bottom',
        line = dict(color = df['hp'],
                   colorscale = 'Electric',
                   showscale = False,
                   cmin = -50000,
                   cmax = 15000),
        dimensions = list([
            dict(range = [1500,7500],
                 constraintrange = [3781,5729],
                 label = '<br> Agricultural <br> Benefits (MM USD)', values = df['ab']),
            dict(range = [6400, 10100],
                 constraintrange = [8338,8453],
                 label = '<br> Hydropower <br> Benefits (MM USD)', values = df['hb']),
            dict(range = [2500,4100],
                 constraintrange = [3373.7,3419.8],
                 label = '<br> Hydropower <br> Production (GWh)', values = df['hp']),
            dict(range = [0.85,0.999],
                 constraintrange = [0.962931,0.975174],
                 label = '<br> Ag. Reliability', values = df['rel']),
            dict(range = [72,44],
                 constraintrange = [55.01,49.96],
                 label = '<br> Ag. Vulnerability (m\u00B3/s)', values = df['vul']),
            dict(range = [0.39, 0.99],
                 constraintrange = [0.575,0.653],
                 label = '<br> Ag. Resilience', values = df['res']),
            dict(range = [21,71],
                 constraintrange = [37.56,44.94],
                 label = '<br> Output Flow (m\u00B3/s)', values = df['out'])])
    ),
)

# fig['layout']['title'] = "Parallel Coordinates Graph of MOEA Results and Averaged Scenarios"
fig.write_image("fig1.pdf")
fig.show()
```
