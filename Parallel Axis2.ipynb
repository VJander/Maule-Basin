```python
import plotly.graph_objects as go

import pandas as pd

df = pd.read_excel('Comparative_Results.xls',sheet_name='Average_Results', usecols='B:J')

fig = go.Figure(data=
    go.Parcoords(
        labelfont=dict(size=9.5),
        tickfont=dict(size=12),
        rangefont=dict(size=12),
        labelside='bottom',
        line = dict(color = df['ab'],
                   colorscale = [[0,'red'],[0.1,'lightseagreen'],[1,'lightseagreen']],
                   showscale = False,
                   cmin = 925,
                   cmax = 100000),
        dimensions = list([
            dict(range = [900,6100],
                 constraintrange = [927,928],
                 label = '<br> Agricultural <br> Benefits (MM USD)', values = df['ab']),
            dict(range = [6640,6940],
                 label = '<br> Hydropower <br> Benefits (MM USD)', values = df['hb']),
            dict(range = [2675,2825],
                 label = '<br> Hydropower <br> Production (GWh)', values = df['hp']),
            dict(range = [0.81,0.99],
                 label = '<br> Ag. Reliability', values = df['rel']),
            dict(range = [71,61],
                 label = '<br> Ag. Vulnerability (m\u00B3/s)', values = df['vul']),
            dict(range = [0.31,0.6],
                 label = '<br> Ag. Resilience', values = df['res']),
            dict(range = [12,36],
                 label = '<br> Output Flow (m\u00B3/s)', values = df['out'])])
    ),
)

# fig['layout']['title'] = "Parallel Coordinates Graph of MOEA Results and Averaged Scenarios"
fig.write_image("fig2.svg")
fig.show()
```
